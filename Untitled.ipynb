```python
# Set input directory, and change working directory - plug in D drive
inDir = "D:\\masters_data\\MODIS_BA" 
os.chdir(inDir)                                                               # Change to working directory
outDir = os.path.normpath(os.path.split(inDir)[0] + os.sep + 'output') + '\\' # Create and set output directory
if not os.path.exists(outDir): os.makedirs(outDir)
```

```python
# select MODIS files
h19_07_MODISFiles = glob.glob('*h19v07*') 
print(len(h19_07_MODISFiles))
print(h19_07_MODISFiles[0])
```

```python
First_h19_07 = h19_07_MODISFiles[:108]
Second_h19_07 = h19_07_MODISFiles[108:]
```

```python
# create BA array for first 9 years (2002 - 2010)
first_BA_1907 = np.zeros((2400, 2400))
NUMBER = 0

for j in First_h19_07:
    filename= First_h19_07[NUMBER]
    dataset = gdal.Open(filename,GA_ReadOnly)
    subdataset = dataset.GetSubDatasets()[1]
    
    with rio.open(First_h19_07[NUMBER]) as dataset:
        crs = dataset.read_crs()
    
    file_name = First_h19_07[NUMBER]
    file = SD(file_name, SDC.READ)   
    
    datasets_dic = file.datasets()
    sds_obj = file.select(0)
    data = sds_obj.get() 
    result = np.where(data > 0)
    
    first_BA_1907[result] += 1
    
    NUMBER += 1
```

```python
# create BA array for second 9 years (2011 - 2018)
second_BA_1907 = np.zeros((2400, 2400))
NUMBER = 0

for j in Second_h19_07:
    filename= Second_h19_07[NUMBER]
    dataset = gdal.Open(filename,GA_ReadOnly)
    subdataset = dataset.GetSubDatasets()[1]
    
    with rio.open(Second_h19_07[NUMBER]) as dataset:
        crs = dataset.read_crs()
    
    file_name = Second_h19_07[NUMBER]
    file = SD(file_name, SDC.READ)   
    
    datasets_dic = file.datasets()
    sds_obj = file.select(0)
    data = sds_obj.get() 
    result = np.where(data > 0)
    
    second_BA_1907[result] += 1
    
    NUMBER += 1
```

```python
Mask_1907 = np.zeros((2400, 2400))
# low_low / low_med / low_high

Mask_1907[((first_BA_1907<1) * (second_BA_1907<1))==1] = 1
Mask_1907[((first_BA_1907<1) * (second_BA_1907>=1) * (second_BA_1907<7))==1] = 2
Mask_1907[((first_BA_1907<1) * (second_BA_1907>=7))==1] = 3

# med_low / med_med / med_high
Mask_1907[((first_BA_1907>=1) * (first_BA_1907<7) * (second_BA_1907<1))==1] = 4
Mask_1907[((first_BA_1907>=1) * (first_BA_1907<7) * (second_BA_1907>=1) * (second_BA_1907<7))==1] = 5
Mask_1907[((first_BA_1907>=1) * (first_BA_1907<7) * (second_BA_1907>=7))==1] = 6

#high_low / high_med / high_high
Mask_1907[((second_BA_1907<1) * (first_BA_1907>=7))==1] = 7
Mask_1907[((first_BA_1907>=7) * (second_BA_1907>=1) * (second_BA_1907<7))==1] = 8
Mask_1907[((first_BA_1907>=7) * (second_BA_1907>=7))==1] = 9

# this takes out where canopy height == 0, but means that the colormaps are hardly visible
# Mask_1907[(canopy_height_array==0)==1]=0
```

```python
change_mask_1907 = np.zeros((2400, 2400))

# decrease
change_mask_1907[Mask_1907 == 4] = 1
change_mask_1907[Mask_1907 == 7] = 1
change_mask_1907[Mask_1907 == 8] = 1

# no change
change_mask_1907[Mask_1907 == 1] = 2
change_mask_1907[Mask_1907 == 5] = 2
change_mask_1907[Mask_1907 == 9] = 2

# no change
change_mask_1907[Mask_1907 == 2] = 3
change_mask_1907[Mask_1907 == 3] = 3
change_mask_1907[Mask_1907 == 6] = 3
```

```python
change_cmap = colors.ListedColormap(["blue", "gainsboro", "red"])
```

```python
# shows where the ICESat-2 data has been
canopy_height_array = pd.read_csv("D:\\masters_data\\output\\csv\\total_canopy_height_array_(meancanopyfile).txt")
fig, axes = plt.subplots() 
img = axes.imshow(canopy_height_array,vmin=0, vmax=1, cmap = "Greens") 
divider = make_axes_locatable(axes) 
cax = divider.append_axes("right", size ="5 %") 
  
color_bar = plt.colorbar(img, cax = cax) 

plt.show()
```

```python
color_map = plt.imshow(change_mask_1907)
plt.title("1907")
img = axes.imshow(change_mask_1907, cmap = change_cmap)
plt.colorbar(img, cax = cax)
color_map.set_cmap(change_cmap)
divider = make_axes_locatable(axes) 
cax = divider.append_axes("right", size ="5 %")
color_bar = plt.colorbar(img, cax = cax) 
plt.show()
```
